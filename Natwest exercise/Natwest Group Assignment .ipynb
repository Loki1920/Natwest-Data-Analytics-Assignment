# In this notebook , I have analyzed the given data by Natwest, perform  EDA(using matplotlib and seaborn) and pandas profiling .


```python
# structure_file_df dataFrame
import pandas as pd

structure_file_path = r"C:\Users\lucky\Machine Learning Projects\Natwest exercise\HealthApp_2k.log_structured.csv"
structure_file_df = pd.read_csv(structure_file_path)

structure_file_df = structure_file_df.drop("LineId",axis = 1)

structure_file_df
```

```python
# templates_df dataFrame
templates_file = "HealthApp_2k.log_templates.csv"
templates_df = pd.read_csv(templates_file)
templates_df
```

```python
# log_df  data 
log_file = "HealthApp_2k.log"
log_columns = ["Time", "Component", "Pid", "Content","EventId","EventTemplate","LineId" ]
log_df = pd.read_csv(log_file, sep='|', header=None, names=log_columns)
log_df = log_df.drop("LineId",axis = 1)
log_df[["EventId","EventTemplate"]] = structure_file_df[["EventId","EventTemplate"]]
log_df


```

```python
# checking for  missing value 
missing_values = log_df.isnull().sum()
missing_values
```

```python
log_df.describe(include='all')
```

```python
# Graphical plot log_df data 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime


# Convert the timestamp column to datetime format
log_df['Time'] = pd.to_datetime(log_df['Time'], format='%Y%m%d-%H:%M:%S:%f')

# Extract date, day of the week, and hour columns
log_df['Date'] = log_df['Time'].dt.date
log_df['DayOfWeek'] = log_df['Time'].dt.day_name()
log_df['Hour'] = log_df['Time'].dt.hour

# Plot the number of steps taken on a daily basis
daily_steps = log_df[log_df['Component'] == 'Step_LSC'].groupby('Date').size()
daily_steps.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Daily Steps')
plt.xlabel('Date')
plt.ylabel('Number of Steps')
plt.show()

# Plot the active times during the day
hourly_activity = log_df.groupby('Hour').size()
hourly_activity.plot(kind='bar', figsize=(10, 6), color='orange')
plt.title('Hourly Activity')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Events')
plt.show()

# Identify workout sessions based on certain events
workout_sessions = log_df[log_df['EventTemplate'].isin(['E42', 'E39'])]
workout_sessions['Duration'] = workout_sessions['Time'].diff().dt.total_seconds()
total_workout_time = workout_sessions['Duration'].sum()
print(f'Total workout time: {total_workout_time} seconds')

# Visualize workout sessions
sns.lineplot(x='Time', y='Duration', data=workout_sessions)
plt.title('Workout Sessions')
plt.xlabel('Time')
plt.ylabel('Duration (seconds)')
plt.show()



```
